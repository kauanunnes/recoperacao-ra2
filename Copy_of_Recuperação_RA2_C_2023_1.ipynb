# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$
A derivada da função é:
$$\frac{df}{dx} = \frac{d}{dx} (x^3) + \frac{d}{dx} (5x^2) + \frac{d}{dx} (9)$$

Para derivar cada termo individualmente, usamos as regras de derivação:
$$ \frac{d}{dx} (x^n) = nx^{x-1} $$ 

Substituindo na função temos:

$$ f′(x)= 3x^2 + 10x − 7 $$

## 2.
$$h(x) = \frac{\ln(x)}{x}$$

A derivada da função é: 
$$ h´(x) = \frac{d}{dx} (\frac{\ln(x)}{x}) $$

Usando a regra do quociente:

$$ \frac{d}{dx} (\frac{u}{v}) = \frac{u′v−uv′}{v^2} $$

Substituindo na derivada, temos:

$$ h´(x) = \frac{\frac{1}{x}*x-ln(x)*1}{x^2} $$

Simplificando:

$$ h´(x) = \frac{1-ln(x)}{x^2} $$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def f_derivative(x):
    return 3*x**2 + 10*x - 7

def h(x):
    return np.log(x) / x

def h_derivative(x):
    return (1 - np.log(x)) / x**2

x = np.linspace(0.1, 5, 100)

y_f = f(x)
y_f_derivative = f_derivative(x)

y_h = h(x)
y_h_derivative = h_derivative(x)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y_f, label='f(x) = $x^3 + 5x^2 - 7x + 9$')
plt.plot(x, y_f_derivative, label="f'(x) = $3x^2 + 10x - 7$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função 1 e sua derivada')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y_h, label='h(x) = $\ln(x)/x$')
plt.plot(x, y_h_derivative, label="h'(x) = $(1 - \ln(x))/x^2$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função 2 e sua derivada')
plt.legend()

plt.tight_layout()
plt.show()
```

##1. 
Calcule a derivada da função: $g(x) = e^x \cos(x)$ para $x=5$.

A derivada da função é:
$$ g´(x) = \frac{d}{dx} (e^x * cos(x)) $$

Para calcularmos no ponto de x = 5, podemos usar a regra do produto e as regras de derivação do exponencial e do cosseno.
Primeiramente utilizando a regra do produto temos:

$$ g´(x) = \frac{d}{dx}(e^x) * cos(x) + e^x * \frac{d}{dx}(cos(x))$$

Sabendo que a derivada do exponencial $ e^x $ é simplesmente $ e^x $, e a derivada do $ cos(x) $ é $ - sin(x) $, podemos substituir essas derivadas, tendo como resultado:

$$ g´(x) = e^x * cos(x) - e^x * sin(x) $$

Calculando para $ x = 5$ , temos:

$$ g´(5) = e^5 * cos(5) - e^5 * sin (5)$$

Calcule a derivada da função: $j(x) = \sqrt{x} \ln(x)$ para $x=3$

A derivada da função é: 

$$ f´(x) = \frac{d}{dx} (\sqrt{x} \ln(x))  $$

Podemos usar a regra do produto e as regras de derivação da raiz quadrada e do logaritmo natural para calcular a derivada no ponto onde $ x = 3$

Aplicando a regra do produto, temos:

$$ f´(x) = \frac{d}{dx} (\sqrt{x}) \ln(x) + \sqrt{x}\frac{d}{dx}ln(x)  $$

A derivada da raiz quadrada $ \sqrt{x} $ pode ser encontrada aplicando a regra do quociente, onde: $$ \frac{d}{dx}(\sqrt{x}) = \frac{1}{2\sqrt{x}}$$

A derivada do logaritmo natural $ ln(x) $ é $\frac{1}{x}$. Substituindo tudo isso, obteremos:

$$ f´(x) = \frac{1}{2\sqrt{x}} ln(x) + \sqrt{x} * \frac{1}{x} $$

Portanto, a derivada da função no ponto onde x = 3 é:

$$ f´(x) = \frac{1}{2\sqrt{3}} ln(3) + \frac{\sqrt{3}}{3} $$


# Questão 3
Usando o python, numpy, encontre o máximo local de $$f(x)=x^2-98x+4$$ para $0 \leq x \leq 100$. 
 

```python
import numpy as np

def f(x):
    return x**2 - 98*x + 4

x = np.linspace(0, 100, 1000)

y = f(x)

max_index = np.argmax(y)

x_max = x[max_index]

y_max = y[max_index]

print("Máximo local encontrado em x =", x_max)
print("Valor máximo local encontrado em f(x) =", y_max)
```
