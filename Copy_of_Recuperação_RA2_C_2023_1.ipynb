# Questão 1
Derivem as funções a seguir. 
Para todas as derivadas apresente a álgebra necessária explicitando matematicamente, com latex, a aplicação de todas as regras, simplificações necessárias. Não deixe de plotar os gráficos das funçoes e das derivadas usando o matplotlib.

## 1.

## 2.

## 1. 
$$f(x) = x^3 + 5x^2 - 7x + 9$$
A derivada da função é:
$$\frac{df}{dx} = \frac{d}{dx} (x^3) + \frac{d}{dx} (5x^2) + \frac{d}{dx} (9)$$

Para derivar cada termo individualmente, usamos as regras de derivação:
$$ \frac{d}{dx} (x^n) = nx^{x-1} $$ 

Substituindo na função temos:

$$ f′(x)= 3x^2 + 10x − 7 $$

## 2.
$$h(x) = \frac{\ln(x)}{x}$$

A derivada da função é: 
$$ h´(x) = \frac{d}{dx} (\frac{\ln(x)}{x}) $$

Usando a regra do quociente:

$$ \frac{d}{dx} (\frac{u}{v}) = \frac{u′v−uv′}{v^2} $$

Substituindo na derivada, temos:

$$ h´(x) = \frac{\frac{1}{x}*x-ln(x)*1}{x^2} $$

Simplificando:

$$ h´(x) = \frac{1-ln(x)}{x^2} $$


```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**3 + 5*x**2 - 7*x + 9

def f_derivative(x):
    return 3*x**2 + 10*x - 7

def h(x):
    return np.log(x) / x

def h_derivative(x):
    return (1 - np.log(x)) / x**2

x = np.linspace(0.1, 5, 100)

y_f = f(x)
y_f_derivative = f_derivative(x)

y_h = h(x)
y_h_derivative = h_derivative(x)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y_f, label='f(x) = $x^3 + 5x^2 - 7x + 9$')
plt.plot(x, y_f_derivative, label="f'(x) = $3x^2 + 10x - 7$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função 1 e sua derivada')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, y_h, label='h(x) = $\ln(x)/x$')
plt.plot(x, y_h_derivative, label="h'(x) = $(1 - \ln(x))/x^2$")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função 2 e sua derivada')
plt.legend()

plt.tight_layout()
plt.show()
```
